## 1. Well-conditioned Quadratic Function

"This notebook runs the experiment for the first benchmark problem: a well-conditioned quadratic function $f(x) = \frac{1}{2}x^{T}Qx$ with $cond(Q) \le 10$ and $dim(x) \ge 10$."

### Cell 1: Imports

```python
import torch  # PyTorch for optimization
import torch.optim as optim  # Built-in optimizers
import numpy as np  # Numerical operations
import matplotlib.pyplot as plt  # Plotting library
from collections import defaultdict  # For storing results
```

### Cell 2: Helper Function - Create Quadratic Function

```python
def create_quadratic_fn(n_dim, cond_num):  # Generates f(x)=0.5*x^T Q x with condition number
    eigenvalues = np.linspace(1, cond_num, n_dim)  # Diagonal eigenvalues
    A = np.random.rand(n_dim, n_dim)  # Random matrix
    P, _ = np.linalg.qr(A)  # Orthogonalize it
    P = torch.tensor(P, dtype=torch.float32)  # Convert to tensor
    D = torch.diag(torch.tensor(eigenvalues, dtype=torch.float32))  # Diagonal matrix
    Q = P @ D @ P.T  # Construct symmetric Q
    def f(x): return 0.5 * x.T @ Q @ x  # Define function
    return f
```

### Cell 3: Helper Function - Run Optimization

```python
def run_function_optimization(func, start_params, optimizer_class, optimizer_params, target_loss, max_iter=20000):  # Run optimizer
    params = start_params.clone().detach().requires_grad_(True)  # Enable gradients
    optimizer = optimizer_class([params], **optimizer_params)  # Create optimizer
    loss_history = []  # Store losses
    for i in range(max_iter):  # Optimization loop
        optimizer.zero_grad()  # Reset grads
        loss = func(params)  # Compute loss
        if loss.item() < target_loss:  # Check target
            loss_history.append(loss.item())
            return i, loss_history
        loss.backward()  # Backprop
        optimizer.step()  # Update
        loss_history.append(loss.item())
    return max_iter, loss_history
```

### Cell 4: Novel Optimizer

```python
class YourNovelOptimizer(optim.Optimizer):  # Custom optimizer
    def __init__(self, params, lr=0.01, **kwargs):  # Initialize
        defaults = dict(lr=lr, **kwargs)
        super(YourNovelOptimizer, self).__init__(params, defaults)
        for group in self.param_groups:
            for p in group['params']:
                state = self.state[p]
                state['step'] = 0  # Track steps

    @torch.no_grad()
    def step(self, closure=None):  # Perform one step
        loss = None
        if closure is not None:
            with torch.enable_grad():
                loss = closure()
        for group in self.param_groups:
            lr = group['lr']
            for p in group['params']:
                if p.grad is None:
                    continue
                grad = p.grad.data  # Gradient
                state = self.state[p]
                state['step'] += 1  # Increment step
                p.data.add_(grad, alpha=-lr)  # Simple SGD step
        return loss
```

### Cell 5: Run Experiment

```python
N_DIM, COND_NUM, TARGET_LOSS, N_RUNS, MAX_ITER = 10, 10, 1e-6, 5, 20000  # Config
optimizers_to_test = {  # Define optimizers
    'SGD_Momentum': (optim.SGD, {'lr': 0.01, 'momentum': 0.9}),
    'Adam': (optim.Adam, {'lr': 0.001, 'betas': (0.9, 0.999)}),
    'RMSprop': (optim.RMSprop, {'lr': 0.01}),
    'YourOptimizer': (YourNovelOptimizer, {'lr': 0.01})
}
results = defaultdict(lambda: defaultdict(list))  # Store results
print(f'Running Well-Conditioned Quadratic (Dim={N_DIM}, Cond={COND_NUM})')
for run in range(N_RUNS):  # Multiple runs
    print(f'--- Run {run+1}/{N_RUNS} ---')
    func = create_quadratic_fn(N_DIM, COND_NUM)  # Create function
    start_params = torch.randn(N_DIM, dtype=torch.float32) * 10  # Init
    for opt_name, (opt_class, opt_params) in optimizers_to_test.items():
        iters, loss_hist = run_function_optimization(func, start_params, opt_class, opt_params, TARGET_LOSS, MAX_ITER)
        results[opt_name]['iterations'].append(iters)
        results[opt_name]['loss_curves'].append(loss_hist)
        print(f'{opt_name} converged in {iters} iterations' if iters < MAX_ITER else f'{opt_name} failed')
print('--- All runs complete ---')
```

### Cell 6: Report Results (Table)

```python
print(f'\n{"="*20} FINAL RESULTS {"="*20}')
print(f'Benchmark 1: Well-Conditioned Quadratic (Target Loss: {TARGET_LOSS})')
print(f'Results over {N_RUNS} runs:\n')
print(f'{'Optimizer':<15} | {'Iterations (Mean ± Std)':<25}')
print('-'*45)
for opt_name in optimizers_to_test.keys():
    iters = np.array(results[opt_name]['iterations'])
    mean_iters, std_iters = np.mean(iters), np.std(iters)
    if mean_iters >= MAX_ITER:
        print(f'{opt_name:<15} | > {MAX_ITER} (Failed)')
    else:
        print(f'{opt_name:<15} | {mean_iters:,.1f} ± {std_iters:,.1f}')
```

### Cell 7: Plot Convergence Curves

```python
plt.figure(figsize=(10,6))  # Plot setup
plt.title(f'Well-Conditioned Quadratic Convergence (Dim={N_DIM}, Cond={COND_NUM})')
plt.xlabel('Iterations')
plt.ylabel('Loss (log scale)')
plt.yscale('log')
for opt_name in optimizers_to_test.keys():  # Loop through optimizers
    curves = results[opt_name]['loss_curves']
    min_len = min(len(c) for c in curves)
    avg_curve = np.mean([c[:min_len] for c in curves], axis=0)
    plt.plot(avg_curve, label=opt_name)
plt.legend()
plt.grid(True, which='both', ls='--')
plt.show()
```
